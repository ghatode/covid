# Import Library 

```python
import pandas as pd
import seaborn as sns
import re
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv('World-19-covid.csv')
```

```python
df.head(5)
```

# Create Pivot Table

```python
df_country = df.pivot_table ('Confirmed',['Date'],'Country/Region')
df_country
```

# Create Correlation Matrix Between Country and Confirmed 

```python
c= df_country.corr().abs()
```

```python
c.head(5)
```

```python
s=c.unstack()
s.head(20)
```

```python
so= s.sort_values(kind="quicksort")
so.head(20)
```

# Find the High Correlation Between The Country and Confirmed

```python
def get_redundant_pairs(c):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = c.columns
    for i in range(0, c.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(c, n=5):
    au_corr = c.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(c)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```

```python
print("Top Absolute Correlations")
print(get_top_abs_correlations(c, 3))
```

# Result 

Colombia and Kenya =0.999982  two Country/Region  have highest correlation coefficient. 

# Data visualization

```python
kot = c[c>=.9]
plt.figure(figsize=(12,8))
sns.heatmap(kot, cmap="Reds")
```
